```python
contact_book = {}
def update(name , number):
    count=0
    for key in contact_book:
            if key == name:
                count+=1
    if count > 0:
        print("Contact already exist\n")
    else:
        contact_book.update({name:number})
def search(name):
    for key in contact_book:
        if key == name:
            print(contact_book[key],"\n")
        else:
            print("No contact found\n")
def delete(name):
    for key in contact_book:
        if key == name:
            contact_book.pop(name)
        else:
            print("no contact found\n")

while True:
    print("1.for update\n")
    print("2.for search\n")
    print("3.for delete\n")
    print("4.for view\n")
    print("5.for exit\n")
    operation = int(input())
    if operation ==1:
        name = input("Enter Name\n")
        number = int(input("enter number\n"))
        if name=="" or name.isnumeric()==True:
            print("Invalid Name\n")
        else:
            update(name , number)
    elif operation == 2:
        name = input("Enter Name\n")
        if name=="" or name.isnumeric()==True:
            print("Invalid Name\n")
        search(name)
    elif operation == 3:
        name = input("Enter Name to delete\n")
        delete(name)
    elif operation == 4:
        for key in contact_book:
            print(key ,  ':' , contact_book[key],"\n")
    elif operation == 5:
        print("exit successfully\n")
        break
    else:
        print("wrong input\n")
        continue
```
